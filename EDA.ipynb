ANÁLISIS DE CORRELACIÓN BIVARIADO

```python
# Calcula la matriz de correlación entre las variables numéricas
matriz_correlacion = df_numericas.corr()

# Crea un mapa de calor utilizando seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
```

GRÁFICO DE DISPERSIÓN PARA VER LA CANTIDAD DE HORAS JUGADAS POR AÑOS

```python
plt.scatter(df_numericas['playtime_forever'], df_numericas['year'], color='green', alpha=0.7)

plt.title('Gráfico de Dispersión')
plt.xlabel('playtime_forever')
plt.ylabel('year')
```

HORAS DE JUEGOS POR CATEGORIA EN CADA AÑO

```python
df.columns
```

```python
genres = ['Action', 'Adventure', 'Animation &amp; Modeling',
       'Audio Production', 'Casual', 'Design &amp; Illustration',
       'Early Access', 'Education', 'Free to Play', 'Indie',
       'Massively Multiplayer', 'Photo Editing', 'RPG', 'Racing', 'Simulation',
       'Software Training', 'Sports', 'Strategy', 'Utilities',
       'Video Production', 'Web Publishing']


relevant_columns = ['year', 'playtime_forever'] + genres

# Filtra el DataFrame según el rango de años deseado
df_filtered = df[(df['year'] >= 2001) & (df['year'] <= 2013)][relevant_columns].copy()

# Agrupa por año y suma las horas jugadas por género
df_grouped = df_filtered.groupby('year')[genres].sum()
df_grouped['total_playtime'] = df_grouped[genres].sum(axis=1)
df_grouped = df_grouped.sort_values('total_playtime', ascending=False)

# Visualiza un gráfico de barras apiladas
df_grouped[genres].plot(kind='bar', stacked=True, figsize=(12, 6))  # Puedes ajustar la paleta de colores aquí
plt.xlabel('Año')
plt.ylabel('Horas Jugadas')
plt.title('Horas Jugadas por Año y Género')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()


```

*2013 registra la mayor cantidad de horas jugadas

TOP JUEGOS CON REVIEWS

```python
# top de títulos por horas jugadas y ordenar de mayor a menor
top_titulos = df.groupby('title')['playtime_forever'].sum().sort_values(ascending=True)

# Limita la cantidad de títulos mostrados para mayor claridad
top_titulos = top_titulos.tail(10)  # Puedes ajustar el número de títulos que deseas mostrar

# Visualiza en un gráfico de barras horizontales con títulos ordenados de mayor a menor
top_titulos.plot(kind='barh', figsize=(12, 6), color='skyblue')  
plt.title('Top Juegos por Horas Jugadas')
plt.xlabel('Horas Jugadas')
plt.ylabel('Títulos')
plt.show()
```

```python
reviews_por_anio = df['posted_year'].value_counts().sort_index()

# Visualiza en un gráfico de barras
reviews_por_anio.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Recuento de Reviews por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Reviews')
plt.show()
```

TOP DESARROLLADORES CON MENOS JUEGOS RECOMENDADOS

```python
# Filtra juegos no recomendados
juegos_no_recomendados = df[df['recommend'] == False]

# Cuenta la cantidad de juegos no recomendados por desarrollador
desarrolladores_menos_recomendados = juegos_no_recomendados['developer'].value_counts()

# Filtra los 10 primeros desarrolladores con menos juegos recomendados
top_desarrolladores = desarrolladores_menos_recomendados.head(10)

# Visualiza en un gráfico de barras
top_desarrolladores.plot(kind='bar', figsize=(12, 6), color='salmon')
plt.title('Top 10 Desarrolladores con Menos Juegos Recomendados')
plt.xlabel('Desarrollador')
plt.ylabel('Número de Juegos No Recomendados')
plt.show()
```

RESEÑAS DE USUARIOS CATEGORIZADAS CON ANÁLISIS DE SENTIMIENTO

```python
# Cuenta el número de reseñas por categoría de sentimiento y año
conteo_sentimientos_por_año = df.groupby(['posted_year', 'review']).size().unstack()

# Visualiza gráfico de barras apiladas
conteo_sentimientos_por_año.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'red', 'blue'])
plt.title('Análisis de Sentimiento de Reseñas de Usuarios por Año')
plt.xlabel('Año')
plt.ylabel('Número de Reseñas')
plt.legend(title='Sentimiento', loc='upper right')
plt.show()
```
