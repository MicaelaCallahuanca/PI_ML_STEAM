```python
#función que lee el archivo json para luego convertirlo en un dataframe
def steam_games():
    df_1 = []
    with gzip.open("C:/Users/micai/OneDrive/Escritorio/PI_ML_STEAM/datasets/steam_games.json.gz", "rt", encoding="utf-8") as file:
        for linea in file:
            try:
                objeto_json = json.loads(linea)
                df_1.append(objeto_json)
            except json.JSONDecodeError as err:   
                pass
    return df_1

df_steam_games = pd.DataFrame(steam_games())
#df_steam_games
```

```python
#elimina los nulos
df_steam_games = df_steam_games.dropna()
df_steam_games
```

```python
#df_steam_games.duplicated() 
#no podemos aplicar el método porque hay ciertas columnas que contienen listas que no son hasheables, o sea son mutables. Necesitamos desanidar
#un valor hash es un valor que representa de manera numérica el contenido del objeto
```

```python
# Convierte la columna 'genres' en variables ficticias/dummy (variable artificiales que representan con valores 0 y 1 a las variables categóricas)
df_steam_games_with_dummies = pd.get_dummies(df_steam_games['genres'].apply(pd.Series).stack()).groupby(level=0).sum().astype(int)

# Resetea el índice para evitar problemas con la concatenación
df_steam_games.reset_index(drop=True, inplace=True)
df_steam_games_with_dummies.reset_index(drop=True, inplace=True)

# Concatena las variables ficticias con el DataFrame original
df_steam_games_with_dummies = pd.concat([df_steam_games, df_steam_games_with_dummies], axis=1)
```

```python
#elimino las columnas que no voy a usar
df_steam_games = df_steam_games_with_dummies.drop(['publisher', 'app_name', 'url', 'tags', 'reviews_url', 'specs', 'price', 'early_access'], axis=1)
```

```python
#convierte a datatime
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'], errors='coerce')

#creo la columna year y extraigo el dato de la columna 'release_date' para ponerlo en la nueva columna
df_steam_games['year'] = df_steam_games['release_date'].dt.year.astype('Int64') 

#elimino la columna 'release_date'
df_steam_games.drop(columns=['release_date'], inplace=True)

#elimino la columna 'genres' porque fue desanidada
df_steam_games.drop(columns=['genres'], inplace=True)
```

```python
#guardo el dataframe limpio en un csv
df_steam_games.to_csv('df_steam_games.csv')
```

```python
#abro el archivo y lo pongo en un df
df_2 = []
with gzip.open('C:/Users/micai/OneDrive/Escritorio/PI_ML_STEAM/datasets/user_reviews.json.gz', 'rt', encoding='utf-8') as f:
    for linea in f:
        df_2.append(eval(linea))
df_reviews = pd.DataFrame(df_2)
df_reviews
```

```python
df_reviews_explode = df_reviews.explode('reviews')
```

```python
df_reviews_concat = pd.concat([df_reviews_explode.drop(['reviews'], axis=1), df_reviews_explode['reviews'].apply(pd.Series)], axis=1)
```

```python
#extrae el año de la columna 'posted', luego se renombra como 'posted_year' y al final se borra la columna 'posted'
df_reviews_concat['posted_year'] = df_reviews_concat['posted'].str.extract(r'(\d{4})').astype('Int64')
df_reviews_concat.drop('posted', axis=1, inplace=True)
```

Sentiment Analysis

```python
# Defino una función para realizar el análisis de sentimiento
def analizar_sentimiento(texto):
    if isinstance(texto, str):
        analysis = TextBlob(texto)
        # Asigno etiquetas de acuerdo con la polaridad del sentimiento
        if analysis.sentiment.polarity < 0:
            return 0  # Sentimiento negativo
        elif analysis.sentiment.polarity == 0:
            return 1  # Sentimiento neutro
        else:
            return 2  # Sentimiento positivo
    else:
        return 1  # Valor si no hay reseña escrita

# Aplica la función a la columna 'review' y crea la columna 'sentiment_analysis'
df_reviews_concat['sentiment_analysis'] = df_reviews_concat['review'].apply(lambda x: analizar_sentimiento(x))

# Reemplaza la columna 'review' con 'sentiment_analysis'
df_reviews_concat.drop('review', axis=1, inplace=True)
df_reviews_concat.rename(columns={'sentiment_analysis': 'review'}, inplace=True)
```

```python
#elimino columnas que no voy a usar
df_reviews_concat.drop([0,'funny', 'last_edited', 'helpful'], axis=1, inplace=True)
```

USER ITEMS

```python
df_3 = []
with gzip.open('C:/Users/micai/OneDrive/Escritorio/PI_ML_STEAM/datasets/users_items.json.gz', 'rt', encoding='utf-8-sig') as f:
    for linea in f:
        df_3.append(eval(linea))
df_users_items = pd.DataFrame(df_3)
```

```python
# json_normalize desanida los datos y crear un nuevo DataFrame
df_users_items = pd.json_normalize(df_3, record_path=['items'], meta=['steam_id', 'items_count', 'user_id', 'user_url'])

# Mueve las columnas de 'meta' al final del DataFrame
columnas_ordenadas = df_users_items.columns.tolist()[4:] + df_users_items.columns.tolist()[:4]
df_users_items = df_users_items[columnas_ordenadas]

# Eliminar la columna 'playtime_2weeks'
df_users_items.drop('playtime_2weeks', axis=1, inplace=True)
```

```python
df_users_items.isnull().sum()   #no tiene nulos
```

```python
#suma los duplicados
df_users_items.duplicated().sum()
```

```python
#borra los duplicados
df_users_items = df_users_items.drop_duplicates()
```

```python
#filtro la columnas 'playtime_forever' para que no me tome en cuenta los valores 0
df_users_items = df_users_items[(df_users_items['playtime_forever'] != 0)]
```

```python
df_users_items = df_users_items.drop('steam_id', axis=1)
```
